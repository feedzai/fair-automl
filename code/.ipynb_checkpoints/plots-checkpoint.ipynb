### A.2. Results processing 

Now that the results are already on their final form, we want to transform them in a way that allow us to represent the result of each HO algorithm.

```python
# Unique identifier of a hyperparameter tuner model (unit of comparison)
ID_COLS = ['eval_experiment_uuid', 'run_uuid', 'index']
```

```python
def metrics_over_budget(run_df: pd.DataFrame, alpha: Union[str, float]) -> pd.DataFrame:
    """Computes the α-selection metrics for a specific run."""
    run_idx = run_df.index[0]
    run_df = run_df.reset_index()
    run_uuid = run_df['run_uuid'].unique()[0]
    assert run_df['run_uuid'].nunique() == 1
    
    data_over_budget = pd.DataFrame(
        columns=[
            'run_uuid',
            'budget',
            'validation_performance', 
            'validation_fairness', 
            'metric_val',
            'test_performance', 
            'test_fairness',
            'alpha',
            'mean_p',
            'mean_f',
        ],
        dtype= float,
    )
    
    # First row:
    data_over_budget = data_over_budget.append({
        'run_uuid': run_uuid,
        'budget': 0,
        'validation_performance': 0,
        'validation_fairness': 0,
        'metric_val': 0,
        'test_performance': 0,
        'test_fairness': 0,
        'alpha': alpha if alpha != 'dynamic' else 0.5,
        'mean_p': 0,
        'mean_f': 0,
    }, ignore_index=True)
    
    # Alias for the metric names in validation
    val_perf_metric = f"validation_{PERFORMANCE_METRIC}"
    val_fair_metric = f"validation_{FAIRNESS_METRIC}"
    
    curr_budget = 0
    for idx, row in run_df.iterrows():
        curr_budget += row.ri
        rows_so_far = run_df.loc[:idx,:]

        if alpha == 'dynamic':
            p = rows_so_far[val_perf_metric].mean()
            f = rows_so_far[val_fair_metric].mean()
            a = 0.5 + (f - p) / 2
        else:
            a = float(alpha)
            p, f = -1, -1
        
        metric_vals = rows_so_far[val_perf_metric] * a + (1 - a) * rows_so_far[val_fair_metric]
        metric_val_max = np.max(metric_vals)
        idx_max = np.argmax(metric_vals)
        
        data_over_budget = data_over_budget.append({
            'run_uuid': run_uuid,
            'budget': curr_budget,
            'validation_performance': rows_so_far.iloc[idx_max][val_perf_metric],
            'validation_fairness': rows_so_far.iloc[idx_max][val_fair_metric],
            'metric_val': metric_val_max,
            'model_uuid': run_df.loc[idx_max, 'model_uuid'],
            'test_performance': rows_so_far.iloc[idx_max][f"test_{PERFORMANCE_METRIC}"],
            'test_fairness': rows_so_far.iloc[idx_max][f"test_{FAIRNESS_METRIC}"],
            'alpha': a,
            'mean_p': p,
            'mean_f': f,
        }, ignore_index=True)

    return data_over_budget


def compute_selection_alpha(data: pd.DataFrame) -> dict:
    """Collects the alpha information for all tuners."""
    from collections import defaultdict
    all_data = defaultdict(list)
    
    tuners = data.tuner.unique()
    for tuner in tuners:
        print(f"Processing {tuner}")
        data_task = data[data["tuner"] == tuner] 
        
        alpha_vals = data_task.alpha.unique()
        for alpha_val in alpha_vals:
            alpha_df = data_task[data_task['alpha'] == alpha_val]
            for r in alpha_df['run_uuid'].unique():
                run_df = alpha_df[alpha_df['run_uuid'] == r]

                data_over_budget = metrics_over_budget(run_df, alpha_val)
                all_data[tuner].append(data_over_budget.reset_index(level=0, inplace=False))
    return all_data


def compute_avg_std(data: dict) -> dict:
    """Compute the avg/std of best metric per run per tuner"""
    lines_df = None
    lines_avg_std = dict()
    for a in data.keys():
        lines_avg_std[a] = pd.concat(data[a])
        lines_avg_std[a]['alpha'] = a
    
        if lines_df is None:
            lines_df = lines_avg_std[a].copy()
        else:
            lines_df = lines_df.append(lines_avg_std[a])

    return lines_avg_std
```

```python
# Changing to naming convention used on the paper
tuner_rename = {
    'HB': 'HB',
    'RS': 'RS',
    'FB (α=0.5)': 'FB-bal',
    'RS (α=0.5)': 'RS-bal',
    'FB': 'FB-auto',
}

all_artifacts['tuner'] = all_artifacts['tuner'].apply(
    lambda tuner: tuner_rename[tuner] if tuner in tuner_rename else tuner
)
```

```python
# Re-compute selection alpha results per tuner (per run)
all_results = compute_selection_alpha(all_artifacts)

# Compute statistical measures (mean and std)
all_lines_avg_std = compute_avg_std(all_results)
```

```python
def add_linestyle(lines, fair_style=FAIR_STYLE, perf_style=PERF_STYLE) -> None:
    """Assumes fairness is always plotted first!"""
    for i, line in enumerate(lines):
        if i % 2 == 0:
            line.set_linestyle(fair_style) # fairness
        else:
            line.set_linestyle(perf_style) # performance

            
def add_custom_legend_lineplot(axes, algorithms) -> None:
    algorithms = list(algorithms) if not algorithms else list(algorithms)
    algorithms = sorted(algorithms)
    color_legend = [Line2D([0], [0], color=algo_colors[a], markersize=2) for a in algorithms], algorithms    
    
    handles, labels = axes.get_legend_handles_labels()
    color_leg = axes.legend(*color_legend, 
                            title = 'HO Algorithm',
                            bbox_to_anchor = (-0.70, -0.25),
                            loc            = 'upper left',
                            ncol           = 5,
                            borderaxespad  = 0.,
                            handletextpad  = 0.1,
                            columnspacing  = 0.1,
                            frameon        = False,
                            labelspacing   = 0,
                            markerscale    = 0.5,
                            numpoints      = 50,
                           )
```

```python
def line_plot(
        algorithm_results, 
        matrix_type: str,
        max_budget: float=2348.200,
        xlim: tuple = (-30, 2348.200),
        ylim: tuple = (-0.03, 1.03),
    ) -> None:
    """Draw Fairness-Predictive Accuracy over budget lineplots."""
    METRICS = [f'{matrix_type}_fairness', f'{matrix_type}_performance']
    LABELS = [FAIRNESS_LABEL, PERFORMANCE_LABEL]
    fig, axes = plt.subplots(1, 2)    
    for i, (metric, label) in enumerate(zip(METRICS, LABELS)):
        fig, ax = fig, axes[i]
        ax.set_title(label)

        for a, data in algorithm_results.items():
            data = data.copy() 
            data['budget'] = data['budget'].clip(0, max_budget)
            #^NOTE:     
            # ---------------------------------------------------------------------
            # Hyperband-based algorithms only use up to 2348.199 budget units. 
            # To perform a fairer comparison between HB and RS we had to consider 
            # 24 model-configs because 23 would only yield 2300 budget units used 
            # and HB would have an additional training units. For that reason, 
            # we ceiled the result and ran 24 models for 100 budget units and 
            # when analysing it, we clip these budget units up to the maximum
            # budget used by any HB iteration (: 
            # ---------------------------------------------------------------------
            sns.lineplot(x='budget', y=metric, data=data, color=algo_colors[a], ci=95, ax=ax)
                
        ax.set_xlim(*xlim)
        ax.set_ylim(*ylim)
        ax.set_ylabel(None)
    
    for ax in axes.flat:
        ax.set(xlabel='Budget', ylabel='Objective')
    
    # Hide x labels and tick labels for top plots and y ticks for right plots.
    for ax in axes.flat:
        ax.label_outer()
    
    sns.despine()
    plt.tight_layout()
    add_custom_legend_lineplot(fig.axes[-1], algorithm_results.keys())

    path = f'{IMGS_DIR}/{matrix_type}_fairness_accuracy_over_budget_horizontal.png'
    plt.gcf().savefig(path, dpi=DPI, bbox_inches="tight")
    return plt.show()
```

### B.2. Plot of Selected Models per HO algorithm

```python
PARETO_FEATURE = 'is_pareto_optimal'
SELECTED_FEATURE = 'is_selected'
TARGET_COLS = ['validation_performance', 'validation_fairness', 'test_performance', 'test_fairness','budget', 'model_uuid']

# Update figure size
FIG_SIZE = (4, 4)
plt.rcParams['figure.figsize'] = FIG_SIZE
```

```python
def get_top_model_static_alpha(run_df, alpha, matrix_type='validation'):
    metric_vals = run_df.apply(
        lambda row: 
            row[f'{matrix_type}_{PERFORMANCE_METRIC}'] * alpha +
            row[f'{matrix_type}_{FAIRNESS_METRIC}'] * (1-alpha),
        axis=1,
    )
    max_idx = np.argmax(metric_vals)
    return run_df.iloc[max_idx]

def get_top_model(run_df, alpha='dynamic', matrix_type='validation'):
    if alpha == 'dynamic':
        avg_f = run_df[f'{matrix_type}_{FAIRNESS_METRIC}'].mean()
        avg_p = run_df[f'{matrix_type}_{PERFORMANCE_METRIC}'].mean()
        alpha = 0.5 + ((avg_f - avg_p) / 2)
        #print(f'Dynamic alpha for run {run_df["run_uuid"].unique()[0]} = {alpha}')
    else:
        alpha = float(alpha)

    assert isinstance(alpha, (float, int))
    return get_top_model_static_alpha(run_df, alpha)
```

```python
selected = all_artifacts.copy()
selected[SELECTED_FEATURE] = False

top_models = selected.groupby('run_uuid').apply(
    lambda df: get_top_model(df, alpha=df['alpha'].unique()[0])
)

for idx, row in top_models.iterrows():
    selected.loc[
        (selected['run_uuid'] == row['run_uuid']) &
        (selected['model_uuid'] == row['model_uuid']),
        SELECTED_FEATURE
    ] = True
```

```python
# Sanity check:
# -> assert that top_models correspond to models selected on the last iteration of metrics_over_budget
for tuner, results in all_results.items():
    for run in results:
        run_uuid = run['run_uuid'].unique()[0]
        assert run['run_uuid'].nunique() == 1

        selected_1 = run.iloc[len(run)-1]['model_uuid']
        selected_2 = top_models[top_models['run_uuid'] == run_uuid]['model_uuid']
        assert len(selected_2) == 1
        selected_2 = selected_2.iloc[0]

        assert selected_1 == selected_2, \
            f"Model selection didn't match for run {run_uuid}: {selected_1} vs {selected_2}"
```

```python
def add_custom_legend(axes, title, values, palette) -> None:
    if isinstance(palette, dict):
        color_legend = [Line2D([0], [0], lw=0, marker='o', color=palette[val], label=val) for val in values], values
    else:
        color_legend = [Line2D([0], [0], lw=0, marker='o', color=palette[i], label=val) for i, val in enumerate(values)], values
        
    handles, labels = axes.get_legend_handles_labels()
    color_leg = axes.legend(*color_legend, 
                            # title=title,
                            bbox_to_anchor = (-0.25, -0.20),
                            loc            = 'upper left',
                            ncol           = 5,
                            borderaxespad  = 0.,
                            handletextpad  = 0.1,
                            columnspacing  = 0.1,
                            frameon        = False,
                            labelspacing   = 0,
                           )

    axes.add_artist(color_leg)

    
def plot_tradeoff(df, matrix_type, pred_col=SELECTED_FEATURE, hue=None, xlim=(-0.01,1.01), ylim=(0, 1.05), palette=sns.color_palette('viridis')):
    df = df.copy()
    perf_metric = f"{matrix_type}_{PERFORMANCE_METRIC}"
    fair_metric = f"{matrix_type}_{FAIRNESS_METRIC}"
    if pred_col == PARETO_FEATURE:
        pred_col = f"{matrix_type}_{pred_col}"
    else:
        pred_col = pred_col
    if hue:
        df[hue] = df[hue].astype(str)
        labels=sorted(df[hue].unique())
        hue_args = {
            "hue": hue,
            "hue_order": labels,
            "palette": palette,
        }
    else: 
        hue_args = {}
    
    col_mask = df[pred_col]
    
    ax = sns.scatterplot(
        x=perf_metric, 
        y=fair_metric,
        data=df[~col_mask],
        size=pred_col,
        size_order=(True, False),
        sizes=(SMALL_POINTS, LARGE_POINTS),
        alpha=SMALL_ALPHA*2,
        **hue_args,
    )
    ax = sns.scatterplot(
        x=perf_metric, 
        y=fair_metric,
        data=df[col_mask],
        size=pred_col,
        size_order=(True, False),
        sizes=(SMALL_POINTS, LARGE_POINTS),
        alpha=1,
        **hue_args,
    )
    
    if hue:
        add_custom_legend(ax.axes, hue, labels, palette)
    
    
    
    # Legend for point sizes
    if DATASET_NAME.lower() == 'adult':
        leg_color = 'grey'
        marker_big = Line2D([0], [0], lw=0, marker='o', color=leg_color, markersize=np.sqrt(LARGE_POINTS))
        marker_small = Line2D([0], [0], lw=0, marker='o', color=leg_color, markersize=np.sqrt(SMALL_POINTS), alpha=SMALL_ALPHA*2)

        size_legend = ax.axes.legend(
            (marker_big, marker_small), ('selected', 'not selected'),
            loc = 'lower left',
            frameon=False,
            handletextpad=0.2, borderpad=0.,
        )
        ax.axes.add_artist(size_legend)
    
    plt.gcf().tight_layout()
    sns.despine()
    # plt.title(fr'Fairness vs. Predictive Accuracy tradeoff (in {matrix_type}, {DATASET_NAME})')
    plt.xlabel(PERFORMANCE_LABEL)
    plt.ylabel(FAIRNESS_LABEL)
    plt.ylim(*ylim)
    plt.xlim(*xlim)
    
    if hue:
        path = f'{IMGS_DIR}/{matrix_type}_fairness_performance_{pred_col}_by_{hue}.png'
    else:
        path = f'{IMGS_DIR}/{matrix_type}_fairness_performance_{pred_col}.png'
    
    
    plt.gcf().savefig(path, dpi=DPI, bbox_inches="tight")
    return plt.show()
```

```python
args ={}

if DATASET_NAME.lower() == 'adult':
    args = { 'xlim': (0.7, 1.01), "ylim": (0.4, 1.01)}

elif DATASET_NAME.lower() == 'compas':
    args = { 'xlim': (0.5, 0.92), 'ylim': (-0.01, 1.01)}

elif DATASET_NAME.lower() == 'donorschoose':
    args = { 'xlim': (0.5, 0.65), 'ylim': (0.1, 1.03)}
elif DATASET_NAME.lower() == 'aof':
    args = { 'xlim': (0, 0.8), "ylim": (0.2, 1.01)}


# Selected models (in validation)
plot_tradeoff(selected, 'validation', hue='tuner', pred_col=SELECTED_FEATURE, palette=algo_colors, **args)
```

### B.3. Heatmap of Pareto optimal models per Fairband iteration

```python
# Update visualization properties
FIG_SIZE = (4, 4)
plt.rcParams['figure.figsize'] = FIG_SIZE
```

```python
def is_pareto_efficient(points: np.ndarray) -> np.ndarray:
    """Determines which points are Pareto efficient (weak definition).

    This definition assumes a minimization problem across all dimensions
    of the provided points. In order to apply this function to a different
    type problem consider using the conversion method ``to_minimization``.

    Parameters
    ----------
    points: np.ndarray[Union[float, int]]
        An (n_points, n_objectives) array.

    to_maximize: List[boolean], optional
        A list with n_objectives boolean values specifying which
        objectives are to be maximized.

    Returns
    -------
    np.ndarray[bool]
        A (n_points,) boolean array, indicating the Pareto efficient
        points.
    """
    is_efficient = np.ones(points.shape[0], dtype=bool)
    for i, point in enumerate(points):
        if is_efficient[i]:
            # Keep any point with a lower cost
            is_efficient[is_efficient] = np.any(
                points[is_efficient] < point, axis=1
            )
            # And keep self
            is_efficient[i] = True
    return is_efficient

def compute_pf(data, matrix_type, groupby_col='eval_experiment_uuid'):
    data = data.copy()
    PARETO_FEATURE_TEMP = f"{matrix_type}_{PARETO_FEATURE}"
    PERF_METRIC_TEMP = f"{matrix_type}_{PERFORMANCE_METRIC}"
    FAIR_METRIC_TEMP = f"{matrix_type}_{FAIRNESS_METRIC}"
    
    # Compute pareto efficiency of each solution
    pareto_efficient = data.groupby(groupby_col).apply(
        lambda df: is_pareto_efficient(1 - df[[PERF_METRIC_TEMP, FAIR_METRIC_TEMP]].to_numpy())
    )

    # Drop if previously added
    data.drop(columns=PARETO_FEATURE_TEMP, inplace=True, errors='ignore')

    # Set the newly computed one 
    data.insert(
        loc=len(data.columns),
        column=PARETO_FEATURE_TEMP,
        value=np.zeros((len(data))) - 1,
    )
    
    for _uuid, _run_sols in pareto_efficient.iteritems():
        hb_run_mask = data[groupby_col] == _uuid
        data.loc[hb_run_mask,PARETO_FEATURE_TEMP] = _run_sols
        
    return data
```

```python
artifacts_per_run = selected.copy()
artifacts_pf_per_run = compute_pf(artifacts_per_run, 'validation', 'run_uuid')
artifacts_pf_per_run[f'validation_{PARETO_FEATURE}'].value_counts()

# We're only considering FB-auto for this analysis
heatmap_artifacts = artifacts_pf_per_run[artifacts_per_run['tuner'] == 'FB-auto']
pareto_feature = f"validation_{PARETO_FEATURE}"

# Count Pareto optimal models as a percentage of total sampled models
pareto_density = heatmap_artifacts.groupby(['si', 'i']).apply(
    lambda df: len(df[df[pareto_feature] == True]) / len(df)
)
pareto_density = pareto_density.unstack()
pareto_density
```

```python
# Set in same format as Hyperband Table!
pareto_density = pareto_density.iloc[::-1].transpose()
pareto_density
```

```python
def add_custom_legend(axes, title, values, palette, **kwargs) -> None:
    if isinstance(palette, dict):
        color_legend = [Line2D([0], [0], lw=0, marker='o', color=palette[val], label=val) for val in values], values
    else:
        color_legend = [Line2D([0], [0], lw=0, marker='o', color=palette[i], label=val) for i, val in enumerate(values)], values

    handles, labels = axes.get_legend_handles_labels()
    '''
    color_leg = axes.legend(*color_legend, 
                            # title=title,
                            bbox_to_anchor = (1.05, 0),
                            loc            = 'lower left',
                            borderaxespad  = 0.,
                            **kwargs
                           )
    '''
    color_leg = axes.legend(
        *color_legend,
        loc           ='upper center',
        ncol          = 6,
        #borderaxespad = 0.,
        bbox_to_anchor = (0.45, -0.2),
        handletextpad=0.3,
        columnspacing=1,
        **kwargs
    )

    axes.add_artist(color_leg)


def plot_results(
        df, matrix_type, hue=None, palette=None,
        xlim=(-0.01,1.01), ylim=(0, 1.05),
        background_points=False,
    ):
    if palette is None:
        palette = sns.color_palette('viridis', len(df[hue].unique()))

    LARGE_POINTS = 30
    SMALL_POINTS = 10
    
    LARGE_ALPHA = 1.
    SMALL_ALPHA = 0.2

    df = df.copy()
    perf_metric = f"{matrix_type}_{PERFORMANCE_METRIC}"
    fair_metric = f"{matrix_type}_{FAIRNESS_METRIC}"
    pred_col = SELECTED_FEATURE
    if hue:
        df[hue] = df[hue].astype(str)
        labels=sorted(df[hue].unique())
        hue_args = {
            "hue": hue,
            "hue_order": labels,
            "palette": palette,
        }
    else: 
        hue_args = {}
    
    col_mask = df[pred_col]

    if background_points:
        ax2 = sns.scatterplot(
            x=perf_metric,
            y=fair_metric,
            data=df[~col_mask],
            alpha=SMALL_ALPHA, s=SMALL_POINTS,
            **hue_args,
        )

        # Legend for point sizes
        leg_color = 'grey'
        marker_big = Line2D([0], [0], lw=0, marker='o', color=leg_color, markersize=np.sqrt(LARGE_POINTS))
        marker_small = Line2D([0], [0], lw=0, marker='o', color=leg_color, markersize=np.sqrt(SMALL_POINTS), alpha=SMALL_ALPHA*2)
        
        size_legend = ax2.axes.legend(
            (marker_big, marker_small), ('selected', 'not selected'),
            loc = 'lower left',
            frameon=False,
            handletextpad=0.2, borderpad=0.,
        )
        ax2.axes.add_artist(size_legend)

    ax = sns.scatterplot(
        x=perf_metric, 
        y=fair_metric,
        data=df[col_mask],
        alpha=0.6 if not background_points else 1,
        s=LARGE_POINTS,
        **hue_args,
    )

    if hue:
        add_custom_legend(ax.axes, hue, labels, palette, frameon=False)
    
    plt.gcf().tight_layout()
    sns.despine()
    # plt.title(fr'Fairness vs. Predictive Accuracy tradeoff (in {matrix_type}, {DATASET_NAME})')
    plt.xlabel(PERFORMANCE_LABEL)
    plt.ylabel(FAIRNESS_LABEL)
    plt.ylim(*ylim)
    plt.xlim(*xlim)

    path = f'{IMGS_DIR}/EG_experiment.png'
    plt.gcf().savefig(path, dpi=DPI, bbox_inches="tight")

    return plt.show()
```

```python
model_palette = sns.color_palette('tab10', df['Model Type'].nunique())
plt.rcParams['figure.figsize'] = (5, 3.5)

plot_results(
    df, 'validation', hue='Model Type',
    palette=model_palette, background_points=True,
    **args,
)
```
